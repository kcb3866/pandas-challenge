```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
purchase_data = "Resources/purchase_data.csv"

```

```python
#view data headers
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

```python
#obtain general data stats
purchase_data_df.describe()
```

```python
#obtain data tyoes
purchase_data_df.dtypes
```

# Player Count

```python
#count number of players
player_count = len(purchase_data_df["SN"].unique())
playercount_df = pd.DataFrame({"Total Players": [player_count]})

playercount_df

```

# Purchasing Analysis (Total) 

```python
#purchasing analysis
unique_items = len(purchase_data_df["Item ID"].unique())
# unique_items
#calculate Price
average_price = purchase_data_df["Price"].mean()
# average_price

#number of purchases
num_of_pur = len(purchase_data_df["Purchase ID"])
# num_of_pur

#calculate total revenues
total_revenues= (purchase_data_df["Price"]).sum()
# total_revenues
#place data ino summary df
summary_table = pd.DataFrame({"Number of unique items": [unique_items], 'Average Price': [average_price],
                        "Number of purchases": [num_of_pur], 'Total Revenues': [total_revenues]})
summary_table = summary_table.round(2)
summary_table['Average Price'] = summary_table['Average Price'].map("${:,.2f}".format)
summary_table['Total Revenues'] = summary_table['Total Revenues'].map("${:,.2f}".format)
summary_table['Number of purchases'] = summary_table['Number of purchases'].map("{:,}".format)

summary_table = summary_table.loc[:, ["Number of unique items",'Average Price', "Number of purchases", 'Total Revenues']]

summary_table
```

#  Gender Demographics

```python
#Gender Demographics
gen_dem = purchase_data_df.loc[:, ["Gender", "SN", "Age"]]
gen_dem_count = gen_dem.drop_duplicates()
# gen_dem_count
#get count
count = gen_dem_count['Gender'].value_counts()
# count
gender_count_pct = count / player_count
# gender_count_pct

gender_demo = pd.DataFrame({'Total Count': count, 'Percentage of Players': gender_count_pct})

gender_demo['Percentage of Players'] = gender_demo['Percentage of Players'].map("{:,.2%}".format)

gender_demo
```

#  Purchasing Analysis (Gender)

```python
#get count
count = gen_dem_count['Gender'].value_counts()

#purchase Count by gen
Gen_purchase_cnt = purchase_data_df.groupby(["Gender"]).count()["Purchase ID"]

#AVerage Price
gen_purchase_price= purchase_data_df.groupby(["Gender"]).mean()["Price"]

#gender total purchase
gen_tot_purchase = purchase_data_df.groupby(["Gender"]).sum()["Price"]

#average purchase
average_tot_person = gen_tot_purchase/count

#purchasing Analysis
gender_data = pd.DataFrame({"Purchase Count": Gen_purchase_cnt,
                                       "Average Purchase Price": gen_purchase_price,
                                      "Total Purchase Value": gen_tot_purchase,
                                      "Average Total Purchase per Person": average_tot_person})

gender_data["Average Purchase Price"] = gender_data["Average Purchase Price"].map("${:,.2f}".format)
gender_data["Total Purchase Value"] = gender_data["Total Purchase Value"].map("${:,.2f}".format)
gender_data ["Purchase Count"] = gender_data["Purchase Count"].map("{:,}".format)
gender_data["Avg Total Purchase per Person"] = gender_data["Average Total Purchase per Person"].map("${:,.2f}".format)
gender_data = gender_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

gender_data
```

## Age Demographics

```python
#establish bin for ages
mybin= [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90,]
#create labels
age_groups = ["<10", '10-14', '20-24','25-29', '30-34', '35-39', '40+']
#create table
purchase_data_df["Age_dem"] = pd.cut(purchase_data_df["Age"], mybin, labels=age_groups)

age_count= purchase_data_df["Age_dem"].value_counts()
age_pct_count = age_count / player_count 

age_demo = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_pct_count})
# age_demo

age_demo["Percentage of Players"] = age_demo["Percentage of Players"].map("{:,.2%}".format)

age_demo.sort_index()


```

#  Purchasing Analysis (Age)

```python
# purchasing Analysis by age

purchase_data_df["Age_dem"] = pd.cut(purchase_data_df["Age"], mybin, labels=age_groups)

# #purchase Count by age
age_purchase_cnt = purchase_data_df.groupby(["Age_dem"]).count()["Price"].rename("Purchase Count")

# #AVerage Price
age_purchase_price= purchase_data_df.groupby(["Age_dem"]).mean()["Price"].rename("Average Purchase Price")

# #gender total purchase
age_tot_purchase = purchase_data_df.groupby(["Age_dem"]).sum()["Price"].rename("Total Purchase Value")

# #average purchase
avg_age_tot_person = age_tot_purchase/age_purchase_cnt

#summary df
age_data = pd.DataFrame({"Purchase Count": age_purchase_cnt,
                                       "Average Purchase Price": age_purchase_price,
                                      "Total Purchase Value": age_tot_purchase,
                                      "Avg Total Purchase per Person": avg_age_tot_person})

age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Total Purchase Value"] = age_data["Total Purchase Value"].map("${:,.2f}".format)
age_data ["Purchase Count"] = age_data["Purchase Count"].map("{:,}".format)
age_data["Avg Total Purchase per Person"] = age_data["Avg Total Purchase per Person"].map("${:,.2f}".format)
age_data = age_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

age_data
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#most Popular item
popular_purchases = purchase_data_df.loc[:,["Item ID", "Item Name", "Price"]]
#Popular item counts

total_item_purchase = popular_purchases.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
average_item_price = popular_purchases.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = popular_purchases.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

# Create new DataFrame
item_data_pd = pd.DataFrame({"Total Purchase Value": total_item_purchase, "Item Price": average_item_price, "Purchase Count": item_count})

# Sort Value
sorted_item_count = item_data_pd.sort_values("Purchase Count", ascending=False)

# Data Wrangling
sorted_item_count["Item Price"] = sorted_item_count["Item Price"].map("${:,.2f}".format)
sorted_item_count["Purchase Count"] = sorted_item_count["Purchase Count"].map("{:,}".format)
sorted_item_count["Total Purchase Value"] = sorted_item_count["Total Purchase Value"].map("${:,.2f}".format)
item_popularity = sorted_item_count.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

item_popularity.head(5)
```

* Sort the above table by total purchase value in descending order

```python
# Item Table (Sorted by Total Purchase Value)
item_total_purchase = item_data_pd.sort_values("Total Purchase Value", ascending=False)

# Minor Data Wrangling
item_total_purchase["Item Price"] = item_total_purchase["Item Price"].map("${:,.2f}".format)
item_total_purchase["Purchase Count"] = item_total_purchase["Purchase Count"].map("{:,}".format)
item_total_purchase["Total Purchase Value"] = item_total_purchase["Total Purchase Value"].map("${:,.2f}".format)
item_profitable = item_total_purchase.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

item_profitable.head(5)
```
