
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Gender Demographics
gen_dem = purchase_data.loc[:, ["Gender", "SN", "Age"]]
gen_dem_count = gen_dem.drop_duplicates()
gen_dem_count
```

```python
#get count
count = gen_dem_count['Gender'].value_counts()

#purchase Count by gen
Gen_purchase_cnt = purchase_data_df.groupby(["Gender"]).count()["Purchase ID"]

#AVerage Price
gen_purchase_price= purchase_data_df.groupby(["Gender"]).mean()["Price"]

#gender total purchase
gen_tot_purchase = purchase_data_df.groupby(["Gender"]).sum()["Price"]

#average purchase
average_tot_person = gen_tot_purchase/count

#purchasing Analysis
purchasing_analysis_df = pd.DataFrame({"Purchase Count": Gen_purchase_cnt,
                                       "Average Purchase Price": gen_purchase_price,
                                      "Total Purchase Value": gen_tot_purchase,
                                      "Average Total per Person": average_tot_person})
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group

```python
#establish bin for ages
mybin= [0, 10, 15, 20, 25, 30, 35, 40]
#create labels
age_groups = ["<10", '10-14', '20-24','25-29', '30-34', '35-39', '40>']
#create table
gen_dem_count["Age_dem"] = pd.cut(gen_dem_count["Age"], bins=mybin, labels=age_groups)

age_count= gen_dem_count.groupby(["Age_dem"]).count()["Age"]
age_count = age_count.to_frame()

age_count["Player Percentage"] = round(age_count["Age"]/player_count, 2)
age_count

#create summary table
# age_summary_df = pd.DataFrame(age_count)
# print(age_summary_df)
# total_player= len(gen_dem_count)
#print(total_player)
# print(type(total_player))
# print(gen_dem_count["Age_dem"])
# gen_dem_count["Player_pct"]= [( (100*gen_dem_count["Age_dem"][i])/(total_player))  for i in np.arange(0,len(gen_dem_count),1)]
# print(gen_dem_count)
```
